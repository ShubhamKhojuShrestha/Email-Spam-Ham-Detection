```python
import pandas as pd
```

<h2>Load Dataset</h2>

```python
df = pd.read_csv('enron_spam_data.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
df["Spam/Ham"].value_counts().plot(kind="bar")
plt.title("Spam vs Ham Distribution")
plt.xlabel("Email Type")
plt.ylabel("Count")
plt.show()
```

<h2>Handling NaN values</h2>

```python
df["Subject"] = df["Subject"].fillna("")
df["Message"] = df["Message"].fillna("")
df = df.dropna(subset=["Spam/Ham"])
df
```

<h2>Combining Text</h2>

```python
df["text"] = df["Subject"] + " " + df["Message"]
df
```

<h2>Coverting Labels</h2>

```python
df["label"] = df["Spam/Ham"].map({"spam": 1, "ham": 0})
df
```

<h2>Pre-processing the texts</h2>

```python
import re
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    return text

df["text"] = df["text"].apply(preprocess_text)
df
```

<h2>Train-Test Split</h2>

```python
from sklearn.model_selection import train_test_split

X_train_text, X_test_text, y_train, y_test = train_test_split( df["text"], df["label"], test_size=0.2, random_state=42)
```

<h2>TF-IDF</h2>

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)

X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)   
```

<h2>Naive Bayes</h2> 

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, f1_score
```

```python
from sklearn.naive_bayes import MultinomialNB

nb_model = MultinomialNB()
nb_model
```

```python
nb_model.fit(X_train, y_train)
```

```python
y_pred_nb = nb_model.predict(X_test)
y_pred_nb
```

```python
nb_accuracy = accuracy_score(y_test, y_pred_nb)
nb_f1 = f1_score(y_test, y_pred_nb)
print("Naive Bayes Accuracy:", nb_accuracy)
print(classification_report(y_test, y_pred_nb))
```

```python
ConfusionMatrixDisplay.from_estimator(nb_model, X_test, y_test, display_labels=["Ham", "Spam"])
plt.title("Naive Bayes – Confusion Matrix") 
plt.show()
```

```python
plt.bar(["Accuracy", "F1-score"], [nb_accuracy, nb_f1])
plt.ylim(0, 1)
plt.title("Naive Bayes Performance")
plt.ylabel("Score")
plt.text(0, nb_accuracy / 2, f"{nb_accuracy:.2f}", ha="center", va="center", color="white")
plt.text(1, nb_f1 / 2, f"{nb_f1:.2f}", ha="center", va="center", color="white")
plt.show()
```

<h2>Logistic Regression</h2>

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model
```

```python
lr_model.fit(X_train, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test)
y_pred_lr
```

```python
lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_f1 = f1_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", lr_accuracy)
print(classification_report(y_test, y_pred_lr))
```

```python
ConfusionMatrixDisplay.from_estimator(lr_model, X_test, y_test,display_labels=["Ham", "Spam"])
plt.title("Logistic Regression – Confusion Matrix")
plt.show()
```

```python
plt.bar(["Accuracy", "F1-score"], [lr_accuracy, lr_f1])
plt.ylim(0, 1)
plt.title("Logistic Regression Performance")
plt.ylabel("Score")
plt.text(0, lr_accuracy / 2, f"{lr_accuracy:.2f}", ha="center", va="center", color="white")
plt.text(1, lr_f1 / 2, f"{lr_f1:.2f}", ha="center", va="center", color="white")
plt.show()
```

<h2>Support Vector Machine (SVM)</h2>

```python
from sklearn.svm import SVC

svm_model = SVC(kernel="linear")
svm_model
```

```python
svm_model.fit(X_train, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)
y_pred_svm
```

```python
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_f1 = f1_score(y_test, y_pred_svm)
print("SVM Accuracy:", svm_accuracy)
print(classification_report(y_test, y_pred_svm))
```

```python
ConfusionMatrixDisplay.from_estimator(svm_model, X_test, y_test, display_labels=["Ham", "Spam"])
plt.title("SVM – Confusion Matrix")
plt.show()
```

```python
plt.bar(["Accuracy", "F1-score"], [svm_accuracy, svm_f1])
plt.ylim(0, 1)
plt.title("SVM Performance")
plt.ylabel("Score")
plt.text(0, svm_accuracy / 2, f"{svm_accuracy:.2f}", ha="center", va="center", color="white")
plt.text(1, svm_f1 / 2, f"{svm_f1:.2f}", ha="center", va="center", color="white")
plt.show()
```

<h2>Cross Validation Check</h2>

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
nb_pipeline = Pipeline([("tfidf", TfidfVectorizer(stop_words="english", max_features=5000)),("nb", MultinomialNB())])
nb_cv_scores = cross_val_score(nb_pipeline, df["text"], df["label"], cv=5, scoring="f1")
print("Naive Bayes CV F1-scores:", nb_cv_scores)
print("Naive Bayes Mean CV F1:", nb_cv_scores.mean())
```

```python
lr_pipeline = Pipeline([("tfidf", TfidfVectorizer(stop_words="english", max_features=5000)),("lr", LogisticRegression(max_iter=1000, C=0.5))])
lr_cv_scores = cross_val_score(lr_pipeline, df["text"],df["label"], cv=5, scoring="f1")
print("Logistic Regression CV F1-scores:", lr_cv_scores)
print("Logistic Regression Mean CV F1:", lr_cv_scores.mean())
```

```python
svm_pipeline = Pipeline([("tfidf", TfidfVectorizer(stop_words="english", max_features=5000)),("svm", SVC(kernel="linear", C=0.5))])
svm_cv_scores = cross_val_score(svm_pipeline, df["text"], df["label"], cv=5, scoring="f1")
print("SVM CV F1-scores:", svm_cv_scores)
print("SVM Mean CV F1:", svm_cv_scores.mean())
```
